```python
import json
import jsonlines
import re
import random

file_path = 'comparison_data/tmp_data/zhihu_3k_rlhf.jsonl'
tmp_list = []
idx = 0
with open(file_path, 'r', encoding="utf-8") as fcc_file:
    for data in jsonlines.Reader(fcc_file):
        norm_d = {'history':data['history'],
                  'query':data['prompt'],
                  'chosen':data['output'][1],
                  'rejected':data['output'][0],
                  'gap_level':0,
                  'id':idx }
        tmp_list.append(norm_d)
        idx += 1

new_file_path = 'comparison_data/zhihu_3k_norm.json'
with open(new_file_path, "w", encoding="utf-8") as dump_f:
    json.dump(tmp_list, dump_f, ensure_ascii=False)
```

```python
import json
import jsonlines
import re
import random

file_path = 'comparison_data/tmp_data/train.jsonl'
tmp_list = []
idx = 0

score = {"风险回复":1, "拒绝为主":3, "拒绝&正向建议":5}

with open(file_path, 'r', encoding="utf-8") as fcc_file:
    for data in jsonlines.Reader(fcc_file):
        norm_d = {'history':[],
                  'query':data['prompt'],
                  'chosen':data['pos_resp'],
                  'rejected':data['neg_resp'],
                  'gap_level':score[data['pos_type']]-score[data['neg_type']],
                  'id':idx }
        tmp_list.append(norm_d)
        idx += 1

new_file_path = 'comparison_data/cvalues_comparison.json'
with open(new_file_path, "w", encoding="utf-8") as dump_f:
    json.dump(tmp_list, dump_f, ensure_ascii=False)
```

```python
import json
data = json.load(open('/data/multiturn_data/knowledge_base_multiturn_new.json', 'rb'))
tmp_list = []
idx = 0

for d in data:
    try:
        norm_d = {'history':d['conversations'][:4],
                  'query':d['conversations'][4]['content'],
                  'id':idx }
        tmp_list.append(norm_d)
        idx += 1
    except:
        continue
    
new_file_path = 'prompt_data/prompt_knowledge_base_multiturn_new.json'
with open(new_file_path, "w", encoding="utf-8") as dump_f:
    json.dump(tmp_list, dump_f, ensure_ascii=False)
```
