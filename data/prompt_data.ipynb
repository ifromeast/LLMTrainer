```python
import json
import random
import jsonlines
import pandas as pd
from tqdm import tqdm

file_name = 'moss_003_mt_data_v2'
mt_data = json.load(open(f'/data/rlhf_data/sft_data_v3/tmp_data/{file_name}.json', 'rb'))

data = random.sample(mt_data, 2000)

idx = 0
tmp_list = []
for d in data:
    rand_len = random.randint(1, 10) * 2
    if len(d['conversations']) <= 6:
        continue
    
    while rand_len >= len(d['conversations'])-2:
        rand_len = random.randint(1, 10) * 2

    norm_d = {'history': d['conversations'][:rand_len],
              'query':d['conversations'][rand_len]['content'],
              'id':idx
             }
    tmp_list.append(norm_d)
    idx += 1

new_file_path = f'{file_name}_2k.json'
with open(new_file_path, "w", encoding="utf-8") as dump_f:
    json.dump(tmp_list, dump_f, ensure_ascii=False)
```

```python
import json
import random
import jsonlines
import pandas as pd
from tqdm import tqdm
from pathlib import Path

dataset_dir = '../sft_data_v3_1'

case_data = []
dataset_path = Path(dataset_dir)
data_files = [file.name for file in dataset_path.glob("*case_*.json")]
for idx, file in enumerate(data_files):
    data_file = os.path.join(dataset_path, file)
    case_data += json.load(open(data_file, 'rb'))

# print(case_data)

case_list = []
idx = 0
for d in case_data:
    length = len(d['conversations']) 
    if length == 2:
        norm_d = {'history':[],
                  'query':d['conversations'][0]['content'],
                  'id':idx }
        case_list.append(norm_d)
        idx += 1
    else:
        try:
            rand_len = random.randint(1, (length-2)//2-1) * 2
        except:
            rand_len = random.randint(1, (length-2)//2) * 2
        # print(rand_len,length)
        norm_d = {'history': d['conversations'][:rand_len],
                  'query':d['conversations'][rand_len]['content'],
                  'id':idx
                 }
        case_list.append(norm_d)
        idx += 1

file_name = 'case_update_prompt'
new_file_path = f'{file_name}.json'
with open(new_file_path, "w", encoding="utf-8") as dump_f:
    json.dump(case_list, dump_f, ensure_ascii=False)
```
